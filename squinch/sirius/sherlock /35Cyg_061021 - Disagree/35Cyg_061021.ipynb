```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pylab as lab
import sys
print ("import done")
```

```python

wave,flux,_ = np.genfromtxt('35Cyg_061021_merged.txt',unpack=True)
```

```python

fig, axes = plt.subplots(2, 1, figsize=(16, 10),dpi=300)
axes[0].set_title("35Cyg_061021 TiO Lines")
axes[0].set_xlabel("Wave") 
axes[0].set_ylabel("Flux") 
axes[0].plot(wave,flux) 
axes[0].set_xlim(400,700)
axes[1].set_title("35Cyg_061021 TiO Lines")
axes[1].set_xlabel("Wave") 
axes[1].set_ylabel("Flux") 
axes[1].plot(wave,flux) 
axes[1].set_xlim(700,1050)
plt.show()

varTioLines =(input("Are TiO lines present (Hint: look for bites missing)? Yes/No"))

if varTioLines == "No" :
    print ("Cool! Now that we figured that out, we need to determine if the Balmer Lines are dominant! Zooming in to 656.3 angstroms...")
if varTioLines == "Yes":
    print ("It's an M type! Done!")
    sys.exit ("End of program. Have a nice day!")
    
```

```python

x = wave
y = flux
lab.title("35Cyg_061021 Balmer Line H Alpha [636,676]")
lab.plot(x, y)
lab.axhline(0, color="black")
lab.axvline(0, color="black")

plt.xlim(636,676)

lab.show()

lab.title("35Cyg_061021 Balmer Line H Alpha [646,666]")
lab.plot(x, y)
lab.axhline(0, color="black")
lab.axvline(0, color="black")

plt.xlim(646,666)

lab.show()

varBalmerLines =(input("Are the Balmer lines dominant? Yes/No"))

if varBalmerLines == "No" :
    print ("Cool! Now that we figured that out, we need to determine if the G-band < H! Zooming in to 434.8 and 430.8 angstroms")
    dominantBalmer = False 
else: 
    if varBalmerLines == "Yes" :
        dominantBalmer = True
        print ("Cool! Now that we figured that out, we need to determine if He is present! Pls proceed to the next cell")

    else:
        sys.exit ("Error, not a valid response. Pls resart program.")
```

```python
if dominantBalmer == True:
    print ("Calssification complete!")
    sys.exit ("Helium is unlikley to to be present. A future update may allow an option to double check this, but for now assume it is an A star! Have a nice day!")
if dominantBalmer == False:
    fig, axes = plt.subplots(2, 1, figsize=(16, 10),dpi=300)
    axes[0].set_title("35Cyg_061021 H 434.8 [429.8,439.8]")
    axes[0].set_xlabel("Wave") 
    axes[0].set_ylabel("Flux") 
    axes[0].plot(wave,flux) 
    axes[0].set_xlim(429.8,439.8)
    axes[1].set_title("35Cyg_061021 G-band 430.8 [425.8,435.8]")
    axes[1].set_xlabel("Wave") 
    axes[1].set_ylabel("Flux") 
    axes[1].plot(wave,flux) 
    axes[1].set_xlim(425.8,435.8)
    plt.show()
   
    
    lab.title("35Cyg_061021 G-band and H")
    lab.plot(x, y)
    lab.axhline(0, color="black")
    lab.axvline(0, color="black")

    plt.xlim(430,436)
    
#extra zoom on G band
    lab.show()
    
    lab.title("35Cyg_061021 G-band 430.8 [429.8,431.8]")
    lab.plot(x, y)
    lab.axhline(0, color="black")
    lab.axvline(0, color="black")

    plt.xlim(429.8,431.8)

    lab.show()
    
    GbandvsH=(input("H (434.8) bigger than G-band (430.8)? (Aka is G-band < H) Yes/No"))
    
    if GbandvsH == "Yes":
        print ("Calssification complete!")
        sys.exit ("Congratulations! It's a F star!")
    if GbandvsH == "No":
        print ("Cool! Now that we figured that out, we need to determine if the H < Fe 432.5! Zooming in to 434.8 and 432.5 angstroms")
```
