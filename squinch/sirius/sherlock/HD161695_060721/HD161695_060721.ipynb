```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pylab as lab
import sys

print ("import done")
```

```python
fileTitle = "HD161695_060721"
wave,flux,_ = np.genfromtxt('HD161695_060721_merged.txt',unpack=True)
```

```python
#TiO line test

x = wave
y = flux

fig, axes = plt.subplots(2, 1, figsize=(16, 10),dpi=300)
axes[0].set_title("TiO Lines (Zoomed Out)")
axes[0].set_xlabel("Wave") 
axes[0].set_ylabel("Flux") 
axes[0].plot(wave,flux)
axes[0].set_xlim(355,1043)
axes[0].axvline(x=1043,color="darkviolet",linewidth=0.5,ls="dashed")
plt.tight_layout()
axes[1].set_title("TiO Lines (Zoomed In)")
axes[1].set_xlabel("Wave") 
axes[1].set_ylabel("Flux") 
axes[1].plot(wave,flux)
axes[1].set_xlim(450,900)
axes[1].axvline(x=1043,color="darkviolet",linewidth=0.5,ls="dashed")
plt.tight_layout()
plt.show()


varTioLines =(input("Are TiO lines present (Hint: look for bites missing)? Yes/No"))

if varTioLines == "No" :
    print ("Cool! Now that we figured that out, we need to determine if the Balmer Lines are dominant! Zooming in to 656.3 angstroms...")
if varTioLines == "Yes":
    print ("It's an M type! Done!")
    sys.exit ("End of program. Have a nice day!")
    
```
