```python
'''This cell imports pandas and the CSV file contiang the data that needs to be sorted and stores it as 'ctf' 

Genral note: This note book was created as a simplified verion of the filter interface notebook and carried over variable names from that project '''

import pandas as pd
import csv
import numpy as np
from astropy import units as u
from astropy.coordinates import Angle



pd.set_option("display.max_rows", None, "display.max_columns", None)

TUI_Targets=pd.read_csv('GeminiTargetsTake1.csv')

display(TUI_Targets)
        
#print ("\nImport Complete\n")
```

```python
def TUI_Convertor(TUI_Targets):
    def Convert_To_DMS(valueOfDec):
        valueOfDec=Angle(valueOfDec, u.degree)
        valueOfDec=valueOfDec.to_string(unit=u.degree,sep=':')
        return valueOfDec

    def Convert_To_HMS(valueOfRA):
        valueOfRA=Angle(valueOfRA, u.degree)
        valueOfRA=valueOfRA.to_string(unit=u.hour, sep=':')
        return valueOfRA

    def MagEquals(GmagVal):
        str(GmagVal)
        GmagVal="Magnitude="+GmagVal.apply(str)
        return GmagVal



    def NameEquals(starName):

        starName='"'+starName+'"'
        return starName






    TUI_Targets2=TUI_Targets

    TUI_Targets2.Dec = Convert_To_DMS(TUI_Targets.Dec)



    TUI_Targets3=TUI_Targets2

    TUI_Targets3.RA = Convert_To_HMS(TUI_Targets.RA)


    TUI_Targets4=TUI_Targets3
    TUI_Targets3.Gmag = MagEquals(TUI_Targets3.Gmag)
    TUI_Targets4.Name = NameEquals(TUI_Targets3.Name)


    TUI_Targets5 = TUI_Targets4.drop(columns=['Class'])
    TUI_TargetsFinal = TUI_Targets5.rename(columns={'Name':'#Name'},index={'One':'Row_1'})


    TUI_TargetsFinal.to_csv('GeminiTargetsTUI_V1.csv',index = False, na_rep="NAN!",sep='\t', quoting=csv.QUOTE_NONE)

    TUI_TargetsFinal.to_csv('GeminiTargetsTUI_V1_NO_TABS.csv',index = False, na_rep="NAN!")

    display(TUI_TargetsFinal)
```

```python
TUI_Convertor(TUI_Targets)
```
