```python
import pandas as pd
import numpy as np
import datetime
import pytz
from astropy.time import Time
from astropy.coordinates import SkyCoord, AltAz, EarthLocation, get_sun
import astropy.units as u
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import astroplan
from astroplan import Observer

# Sample DataFrame with Name, RA, Dec, Class, Gmag columns

df = tsi

# Observatory location
latitude = -29.0146
longitude = -70.6926
elevation = 2380
location = EarthLocation(lat=latitude * u.deg, lon=longitude * u.deg, height=elevation * u.m)

# Date and time range for the night
date = '2023-04-10'
date_next = '2023-04-11'



def get_sunrise_sunset(date, location):
    observer = Observer(location=location)
    
    local_timezone = observer.timezone
    naive_midnight = datetime.datetime.strptime(date, "%Y-%m-%d") + datetime.timedelta(hours=12)
    local_midnight = local_timezone.localize(naive_midnight)
    midnight = Time(local_midnight)
    
    sunset = observer.sun_set_time(midnight, which='nearest')
    sunrise = observer.sun_rise_time(midnight, which='next')

    return sunrise, sunset

sunrise, sunset = get_sunrise_sunset(date, location)
print (sunrise, sunset)

delta_t = sunrise - sunset
timesteps = 10000
time_steps = sunset + delta_t * np.linspace(0, 1, timesteps)

print(time_steps)

# Air mass thresholds
air_mass_thresholds = [3, 2, 1.5]

# Function to calculate the observability of objects at different air mass thresholds
def observability(df, time_steps, location, air_mass_thresholds):
    observable_hours = {threshold: [] for threshold in air_mass_thresholds}

    for index, row in df.iterrows():
        coord = SkyCoord(ra=row['RA'] * u.deg, dec=row['Dec'] * u.deg)
        alt_az = coord.transform_to(AltAz(obstime=time_steps, location=location))
        air_masses = alt_az.secz

        for threshold in air_mass_thresholds:
            observable_time = time_steps[np.where(air_masses <= threshold)]
            observable_duration = (observable_time[-1] - observable_time[0]).to(u.hour) if len(observable_time) > 0 else 0 * u.hour
            observable_hours[threshold].append(observable_duration.value)

    return observable_hours

# Compute observability
observable_hours = observability(df, time_steps, location, air_mass_thresholds)

# Create DataFrame with Name, hours observable at Air Mass < 3, < 2, and < 1.5
observable_df = df[['Name']].copy()
for threshold in air_mass_thresholds:
    observable_df[f"Hours Observable (Air Mass < {threshold})"] = pd.Series(observable_hours[threshold])

print("\nObservable objects and their hours observable at different air mass thresholds:")
display(observable_df)

# Plot observability
fig, ax = plt.subplots()

for index, row in df.iterrows():
    coord = SkyCoord(ra=row['RA'] * u.deg, dec=row['Dec'] * u.deg)
    alt_az = coord.transform_to(AltAz(obstime=time_steps, location=location))
    air_masses = alt_az.secz
    ax.plot(time_steps.plot_date, air_masses, label=row['Name'])

ax.invert_yaxis()
ax.set_ylim([3, 1])
ax.set_xlabel('Time')
ax.set_ylabel('Air Mass')

# Format x-axis
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
plt.xticks(rotation=45)
plt.gcf().subplots_adjust(bottom=0.2)

plt.show()
```
