```python
plt.title("Sample 4") 
plt.xlabel("Wave") 
plt.ylabel("Flux") 
plt.plot(wave1,flux1) 
#plt.xlim(340,700)
plt.show()

plt.title("Sample 5") 
plt.xlabel("Wave") 
plt.ylabel("Flux") 
plt.plot(wave2,flux2) 
#plt.xlim(340,700)
plt.show()
plt.title("V367Cyg") 
plt.xlabel("Wave") 
plt.ylabel("Flux") 
plt.plot(wave3,flux3) 
#plt.xlim(340,700)
plt.show()
```
