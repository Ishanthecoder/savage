```python
def filterFloatRange(catFloatInput,ctfFloatInput):
    global ctfNewFloat
    print (categoryIntro)
    exitToken = False 
    time.sleep(0.50)
    minFloatValInput = (input("Enter the lower bound for the range you want to filter for. If you dont want to set a lower bound enter 'NONE':\n"))
    maxFloatValInput = (input("Enter the upper bound for the range you want to filter for. If you dont want to set a lower bound enter 'NONE':\n"))
    
    placeholderMax = 1000**10
    placeholderMin = -1000**10
    
    if minFloatValInput == "None" or minFloatValInput == "none":
        minFloatValInput = "NONE"
        
    if maxFloatValInput == "None" or maxFloatValInput == "none":
        maxFloatValInput = "NONE"
        
    
    if minFloatValInput == "NONE" and maxFloatValInput == "NONE":
        exitToken = True
        ctfNewFloat = ctfFloatInput
        printInputs = False
        print ("Not filtering values based on", catFloatInput)
        
    if minFloatValInput == "NONE" and maxFloatValInput != "NONE":
        minFloatVal = placeholderMin
        maxFloatVal = float(maxFloatValInput)
        printInputs = False
        print ("Filering for all values in ",catFloatInput," that are less than",maxFloatVal)
        
        
    if minFloatValInput != "NONE" and maxFloatValInput == "NONE":
        minFloatVal = float(minFloatValInput)
        maxFloatVal = placeholderMax
        printInputs = False
        print ("Filering for all values in ",catFloatInput," that are greater than ",minFloatVal)
        
    if minFloatValInput != "NONE" and maxFloatValInput != "NONE":
        minFloatVal = float(minFloatValInput)
        maxFloatVal = float(maxFloatValInput)
        printInputs = True
        print ("Filering for all values in ",catFloatInput," that are greater than ",minFloatVal," and less than",maxFloatVal)
    
        
    if catFloatInput == "RA" and exitToken == False:
        placeholderMin = ctfFloatInput.RA.min()
        placeholderMax = ctfFloatInput.RA.max()
        ctfFloatMin = (ctfFloatInput.RA >= minFloatVal)
        targetStarsIndex = [*filter(ctfFloatMin.get, ctfFloatMin.index)]
        ctfFloatProcessed = ctfFloatInput.loc[targetStarsIndex]
        ctfFloat = (ctfFloatProcessed.RA <= maxFloatVal)
        
    if catFloatInput == "Dec" and exitToken == False:
        placeholderMin = ctfFloatInput.Dec.min()
        placeholderMax = ctfFloatInput.Dec.max()
        ctfFloatMin = (ctfFloatInput.Dec >= minFloatVal)
        targetStarsIndex = [*filter(ctfFloatMin.get, ctfFloatMin.index)]
        ctfFloatProcessed = ctfFloatInput.loc[targetStarsIndex]
        ctfFloat = (ctfFloatProcessed.Dec <= maxFloatVal)
        
        
    if catFloatInput == "Gmag" and exitToken == False:
        placeholderMin = ctfFloatInput.Gmag.min()
        placeholderMax = ctfFloatInput.Gmag.max()
        ctfFloatMin = (ctfFloatInput.Gmag >= minFloatVal)
        targetStarsIndex = [*filter(ctfFloatMin.get, ctfFloatMin.index)]
        ctfFloatProcessed = ctfFloatInput.loc[targetStarsIndex]
        ctfFloat = (ctfFloatProcessed.Gmag <= maxFloatVal)
        
    if exitToken == False:
        targetStarsIndex = [*filter(ctfFloat.get, ctfFloat.index)]
        ctfNewFloat = ctfFloatInput.loc[targetStarsIndex]
    return ctfNewFloat
```

```python
def convertedConvertor(convertedTargets):
    def Convert_To_DMS(valueOfDec):
        valueOfDec=Angle(valueOfDec, u.degree)
        valueOfDec=valueOfDec.to_string(unit=u.degree,sep=':')
        return valueOfDec

    def Convert_To_HMS(valueOfRA):
        valueOfRA=Angle(valueOfRA, u.degree)
        valueOfRA=valueOfRA.to_string(unit=u.hour, sep=':')
        return valueOfRA

    convertedTargets2=convertedTargets

    convertedTargets2.Dec = Convert_To_DMS(convertedTargets.Dec)



    convertedTargets3=convertedTargets2

    convertedTargets3.RA = Convert_To_HMS(convertedTargets.RA)

    global convertedTargetsFinal
    
    convertedTargetsFinal = convertedTargets3

    display(convertedTargetsFinal)
    return convertedTargetsFinal
```

```python
def TUI_Convertor(TUI_Targets):
    def Convert_To_DMS(valueOfDec):
        valueOfDec=Angle(valueOfDec, u.degree)
        valueOfDec=valueOfDec.to_string(unit=u.degree,sep=':')
        return valueOfDec

    def Convert_To_HMS(valueOfRA):
        valueOfRA=Angle(valueOfRA, u.degree)
        valueOfRA=valueOfRA.to_string(unit=u.hour, sep=':')
        return valueOfRA

    def MagEquals(GmagVal):
        str(GmagVal)
        GmagVal="Magnitude="+GmagVal.apply(str)
        return GmagVal



    def NameEquals(starName):

        starName='"'+starName+'"'
        return starName






    TUI_Targets2=TUI_Targets

    TUI_Targets2.Dec = Convert_To_DMS(TUI_Targets.Dec)



    TUI_Targets3=TUI_Targets2

    TUI_Targets3.RA = Convert_To_HMS(TUI_Targets.RA)


    TUI_Targets4=TUI_Targets3
    TUI_Targets3.Gmag = MagEquals(TUI_Targets3.Gmag)
    TUI_Targets4.Name = NameEquals(TUI_Targets3.Name)


    TUI_Targets5 = TUI_Targets4.drop(columns=['Class'])
    global TUI_TargetsFinal
    TUI_TargetsFinal = TUI_Targets5.rename(columns={'Name':'#Name'},index={'One':'Row_1'})

    display(TUI_TargetsFinal)
    
    return TUI_TargetsFinal
    
```
