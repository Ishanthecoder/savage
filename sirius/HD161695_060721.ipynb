```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pylab as lab
import sys
print ("import done")
```

```python

wave,flux,_ = np.genfromtxt('HD161695_060721_merged.txt',unpack=True)
```

```python
plt.title("35Cyg_061021 TiO Lines") 
plt.xlabel("Wave") 
plt.ylabel("Flux") 
plt.plot(wave,flux) 

plt.show()

varTioLines =(input("Are TiO lines present (Hint: look for bites missing)? Yes/No"))

if varTioLines == "No" :
    print ("Cool! Now that we figured that out, we need to determine if the Balmer Lines are dominant! Zooming in to 656.3 angstroms...")
if varTioLines == "Yes":
    print ("It's an M type! Done!")
    sys.exit ("End of program. Have a nice day!")
    
```

```python
fig, axes = plt.subplots(4, 1, figsize=(16, 10),dpi=300)
axes[0].set_title("35Cyg_061021 Calcium Triplet (All)")
axes[0].set_xlabel("Wave") 
axes[0].set_ylabel("Flux") 
axes[0].plot(wave,flux) 
axes[0].set_xlim(834.8,881.2)
axes[1].set_title("Calcium Triplet (849.8)")
axes[1].set_xlabel("Wave") 
axes[1].set_ylabel("Flux") 
axes[1].plot(wave,flux) 
axes[1].set_xlim(844.8,854.8)
axes[2].set_title("Calcium Triplet (854.2)")
axes[2].set_xlabel("Wave") 
axes[2].set_ylabel("Flux") 
axes[2].plot(wave,flux) 
axes[2].set_xlim(849.2,859.2)
axes[3].set_title("Calcium Triplet(866.2)")
axes[3].set_xlabel("Wave") 
axes[3].set_ylabel("Flux") 
axes[3].plot(wave,flux) 
axes[3].set_xlim(861.2,871.2)
plt.show()

varCaTriplet =(input("Are the Calcium Triplet present? Yes/No"))

if varCaTriplet == "Yes" :
    print ("Alright, it's probably a Cool star so expect balmer lines not to be dominant")
    dominantBalmer = False 
else: 
    if varCaTriplet == "No" :
        dominantBalmer = True
        print ("Hmmm, in that case this likley a hot star in which case balmer lines are likey dominant")

    else:
        sys.exit ("Error, not a valid response. Pls resart program.")
```

```python

x = wave
y = flux

fig, axes = plt.subplots(3, 1, figsize=(16, 10),dpi=300)
axes[0].set_title("Balmer Lines (Zoomed Out)")
axes[0].set_xlabel("Wave") 
axes[0].set_ylabel("Flux") 
axes[0].plot(wave,flux) 
axes[0].set_xlim(616,696)
axes[0].axvline(x=656.3,color="darkviolet",linewidth=0.5,ls="dashed")
axes[1].set_title("Balmer Line H Alpha [636,676]")
axes[1].set_xlabel("Wave") 
axes[1].set_ylabel("Flux") 
axes[1].plot(wave,flux) 
axes[1].set_xlim(636,676)
axes[1].axvline(x=656.3,color="darkviolet",linewidth=0.5,ls="dashed")
axes[2].set_title("Balmer Line H Alpha [646,666]")
axes[2].set_xlabel("Wave") 
axes[2].set_ylabel("Flux") 
axes[2].plot(wave,flux) 
axes[2].axvline(x=656.3,color="darkviolet",linewidth=0.5,ls="dashed")
axes[2].set_xlim(646,666)
plt.tight_layout()
plt.show()

varBalmerLines =(input("Are the Balmer lines dominant? Yes/No"))

if varBalmerLines == "No" :
    print ("Cool! Now that we figured that out, we need to determine if the G-band < Hγ! Zooming in to 434.8 and 430.8 angstroms")
    dominantBalmer = False 
else: 
    if varBalmerLines == "Yes" :
        dominantBalmer = True
        print ("Cool! Now that we figured that out, we need to determine if He is present! Pls proceed to the next cell")

    else:
        sys.exit ("Error, not a valid response. Pls resart program.")
        
        
if dominantBalmer == True:
    print ("Calssification complete!")
    sys.exit ("Helium is unlikley to to be present. A future update may allow an option to double check this, but for now assume it is an A star! Have a nice day!")
```

```python
if dominantBalmer == False:
    x = wave
    y = flux
    lab.title("35Cyg_061021 Hγ [425,445]")
    lab.plot(x, y)
    lab.axhline(0, color="black")
    lab.axvline(0, color="black")

    plt.xlim(425,445)

    lab.show()

    lab.title("35Cyg_061021 G-band [421,441]")
    lab.plot(x, y)
    lab.axhline(0, color="black")
    lab.axvline(0, color="black")

    plt.xlim(421,441)

    lab.show()
    
    lab.title("35Cyg_061021 G-band and Hγ")
    lab.plot(x, y)
    lab.axhline(0, color="black")
    lab.axvline(0, color="black")

    plt.xlim(430,436)
    
#extra zoom on G band
    lab.show()
    
    lab.title("35Cyg_061021 G-band [430,432]")
    lab.plot(x, y)
    lab.axhline(0, color="black")
    lab.axvline(0, color="black")

    plt.xlim(429.8,431.8)

    lab.show()
    
    GbandvsHγ=(input("Hγ (434.8) bigger than G-band (430.8)? (Aka is G-band < Hγ) Yes/No"))
    
    if GbandvsHγ == "Yes":
        print ("Calssification complete!")
        sys.exit ("Congratulations! It's a F star!")
    if GbandvsHγ == "No":
        print ("Cool! Now that we figured that out, we need to determine if the Hγ < Fe 432.5! Zooming in to 434.8 and 432.5 angstroms")
```
