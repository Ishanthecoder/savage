```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pylab as lab
print ("import done")
```

```python

wave,flux,_ = np.genfromtxt('StarBlindSample1.txt',unpack=True)
```

```python
plt.title("Test Plot") 
plt.xlabel("Wave") 
plt.ylabel("Flux") 
plt.plot(wave,flux) 
plt.xlim(400,700)
plt.show()
```

```python

x = wave
y = flux
lab.plot(x, y)
lab.axhline(0, color="black")
lab.axvline(0, color="black")
lab.ylim(0.0,0.6)
lab.xlim(410,450)


lab.show()
```
